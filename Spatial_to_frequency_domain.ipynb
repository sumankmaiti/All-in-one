```python
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.color import rgb2hsv, rgb2gray, rgb2yuv
from skimage import color, exposure, transform
from skimage.exposure import equalize_hist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train/255.0
x_test = x_test/255.0
# Plot image data from x_train
test_image = x_test[1].reshape(28, 28)
imshow(test_image)
```

```python
frequency_domain = np.fft.fft2(test_image)
frequency_domain
```

```python
shifted_frequency_domain = np.fft.fftshift(frequency_domain)
shifted_frequency_domain
```

```python
magnitude_of_frequency = abs(frequency_domain)
magnitude_of_frequency
```

```python
plt.imshow(magnitude_of_frequency, cmap='gray')
print("Frequency Domain")
plt.show()
```

```python
magnitude_of_shifted_frequency = abs(shifted_frequency_domain)
magnitude_of_shifted_frequency
```

```python
plt.imshow(magnitude_of_shifted_frequency, cmap='gray')
print("Shifted Frequency Domain")
plt.show()
```
