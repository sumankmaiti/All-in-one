```python
from sklearn.datasets import load_sample_image
import matplotlib.pyplot as plt
```

```python
x= load_sample_image("china.jpg")/255
plt.imshow(x)
x
```

```python
y= load_sample_image("flower.jpg")/255
plt.imshow(y)
```

```python
import numpy as np

images = np.array([x, y])
```

```python
batch, h, w, c = images.shape
print(batch, h, w, c)
```

```python
filters = np.zeros(shape=(7,7,c, 2), dtype= np.float32)
filters
```

```python
filters[:, 3, :, 0] = 1
filters[3, :, :, 1] = 1
```

```python
import tensorflow as tf

outputs = tf.nn.conv2d(images, filters, strides = 1, padding ="SAME")
plt.imshow(outputs[1,:,:,0], cmap ="gray")
```
