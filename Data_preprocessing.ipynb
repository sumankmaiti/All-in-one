signature generation function from dataset

```
from pandas import DataFrame
def signature_generation(dataset):
  List = dataset.values.tolist()  # convert the dataframe to list
  # len(List)
  total_signature = [] # hold the total signatures from a data frame
  for item in List:   # take one row at a time
    ###### convert the values to float##########
    converted = []
    for i in range(len(item)):
      x = float(item[i])
      converted.append(x)
    ####### create bins ################
    converted.sort()  
    min = converted[0]
    # print(min)
    max = converted[len(converted)-1]
    # print(max)
    bin = (max - min)/300
    # print(bin)
    ######## creating bin size ###########
    signature = []
    while min < max:
      # min = min
      upper = (min + bin)
      count = 0
      ########## count the frequency of values #############
      for i in range(len(converted)):
        comp = converted[i]
        if min <= comp <= upper:
          count += 1
      signature.append(count)
      min = upper
    ######## signature using frequency count #########
    total_signature.append(signature)
  # print(total_signature)
  print("length of the signature set", len(total_signature))
  ##### generating data frame from signature list ###############
  device_SIGNATURE = DataFrame(total_signature)
  device_SIGNATURE = device_SIGNATURE.fillna(0)  ## not available values filled with 0
  return device_SIGNATURE
```

```
# encoding catagorical data
from sklearn.preprocessing import LabelEncoder
labelencoder_X_1 = LabelEncoder()
Y1 = labelencoder_X_1.fit_transform(Y)
Y = pd.get_dummies(Y1).values

# splitting the dataset into the training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state = 0)

# feature scalling
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# X_train = sc.fit_transform(X_train)
# X_test = sc.fit_transform(X_test)
```

```
model = Sequential()
model.add(Dense(50, input_shape = (722,), activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))
```

```
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```
model.fit(X_train, Y_train, epochs=100)
```

```
y_pred = model.predict(X_test)
y_test_class = np.argmax(Y_test, axis = 1)
y_pred_class = np.argmax(y_pred, axis = 1)
```

```
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test_class, y_pred_class))
print(confusion_matrix(y_test_class, y_pred_class))
```
